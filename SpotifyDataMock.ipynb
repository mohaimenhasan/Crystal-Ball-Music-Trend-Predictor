```python
# Calculating Eigenvectors and eigenvalues of Cov matirx
from sklearn.preprocessing import StandardScaler
MockData_std = StandardScaler().fit_transform(MockData)
mean_vec = np.mean(MockData_std, axis=0)
cov_mat = np.cov(MockData_std.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
```

```python
#make list of (eigenvalue, eigenvector) tuples
eig_pairs = [ (np.abs(eig_vals[i]),eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort from high to low
eig_pairs.sort(key = lambda x: x[0], reverse= True)

# Calculation of Explained Variance from the eigenvalues
tot = sum(eig_vals)
var_exp = [(i/tot)*100 for i in sorted(eig_vals, reverse=True)] # Individual explained variance
cum_var_exp = np.cumsum(var_exp) # Cumulative explained variance
len(cum_var_exp)
#len(var_exp)
```

```python
# PLOT OUT THE EXPLAINED VARIANCES SUPERIMPOSED 
plt.figure(figsize=(10, 5))
plt.bar(range(13), var_exp, alpha=0.3333, align='center', label='individual explained variance', color = 'g')
plt.step(range(13), cum_var_exp, where='mid',label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.show()
```

```python
# split PCA scores into training and testing set
TrainPCA = MockData_10d[:int(len(MockData_10d) *  0.6)]
TestPCA = MockData_10d[int(len(MockData_10d) *  0.6):]
TrainPCA.shape, X_train.shape, TestPCA.shape, Y_train.shape
```

```python
# find R^2 Score using various regression methods on PCA sets
def lets_try(train,labels):
    results={}
    def test_model(clf):
        
        cv = KFold(n_splits=5,shuffle=True,random_state=45)
        r2 = make_scorer(r2_score)
        r2_val_score = cross_val_score(clf, train, labels, cv=cv,scoring=r2)
        scores=[r2_val_score.mean()]
        return scores

    clf = linear_model.LinearRegression()
    results["Linear"]=test_model(clf)
    
    clf = linear_model.Ridge()
    results["Ridge"]=test_model(clf)
    
    clf = linear_model.BayesianRidge()
    results["Bayesian Ridge"]=test_model(clf)
    
    clf = linear_model.HuberRegressor()
    results["Hubber"]=test_model(clf)
    
    clf = linear_model.Lasso(alpha=1e-4)
    results["Lasso"]=test_model(clf)
    
    clf = BaggingRegressor()
    results["Bagging"]=test_model(clf)
    
    clf = RandomForestRegressor()
    results["RandomForest"]=test_model(clf)
    
    clf = AdaBoostRegressor()
    results["AdaBoost"]=test_model(clf)
    
    clf = svm.SVR()
    results["SVM RBF"]=test_model(clf)
    
    clf = svm.SVR(kernel="linear")
    results["SVM Linear"]=test_model(clf)
    
    results = pd.DataFrame.from_dict(results,orient='index')
    results.columns=["R Square Score"] 
    results=results.sort_values(by='R Square Score', ascending = False)
    results.plot(kind="bar",title="Model Scores")
    axes = plt.gca()
    axes.set_ylim([0.5,1])
    return results

lets_try(TrainPCA, Y_train)
```

```python
cv = KFold(n_splits=5,shuffle=True,random_state=45)

parameters = {'alpha_1': [1e-04, 1e-05, 1e-06,1e-07],
              'alpha_2': [1e-04, 1e-05, 1e-06,1e-07],
              'lambda_1' : [1e-04, 1e-05, 1e-06,1e-07],
              'lambda_2' : [1e-04, 1e-05, 1e-06,1e-07],
              'tol' : [0.001]}

clf = linear_model.BayesianRidge()
r2 = make_scorer(r2_score)
grid_obj = GridSearchCV(clf, parameters,  cv=cv,scoring=r2)
grid_fit = grid_obj.fit(TrainPCA, Y_train)
best_clf = grid_fit.best_estimator_ 

#inputting PCA training set
best_clf.fit(TrainPCA, Y_train)
best_clf.predict(TestPCA)
```

```python
from sklearn.metrics import mean_absolute_error
Y_test_pred = best_clf.predict(TestPCA)
#Y_test_pred = Y_test_pred.astype(int)
Y_test_true = test_df['popularity']
MAE = mean_absolute_error(Y_test_true, Y_test_pred)

from sklearn.metrics import mean_squared_error
RMSE = mean_squared_error(Y_test_true, Y_test_pred)**0.5

#In average, the model has prediction error of ~1.2 unit.
# i.e. the predictred popularity has room of error of (y-1.2, y+1.2)
MAE, RMSE
```
